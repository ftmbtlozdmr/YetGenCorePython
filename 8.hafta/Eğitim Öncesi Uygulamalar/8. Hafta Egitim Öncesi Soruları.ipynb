# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
file = open("urunler.txt", "a", encoding="utf-8")
#r	Mevcut dosyayı okumak için açar, yoksa hata verir
#w	Dosya yoksa oluşturur, varsa içeriği silerek sıfır dosya açar
#a	Dosya yoksa oluşturur, varsa dosyanın sonunda imleçten başlayarak ekleme yapar
#r+	Dosyayı hem okuma, hem yazma(w) modunda açar
def urunBilgisi():
    ad = input("Ürün adını girin: ")
    fiyat = input("Ürün fiyatını girin: ")
    file.write(f"{ad}: {fiyat}TL\n")

urunBilgisi()
file.close()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def isimDegistirme():
    eskiKelime = input("Değiştireceğiniz kelimeyi giriniz: ")
    yeniKelime = input("Kendisiyle değiştirilmesini istediğiniz kelimeyi giriniz: ")
    
    with open("urunler.txt", "r+", encoding="utf-8") as file:
        icerik = file.read()
        icerik = icerik.replace(eskiKelime, yeniKelime)
        file.seek(0)
        file.write(icerik)
        
isimDegistirme()
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class vehicle:
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
class bus(vehicle):
    pass

a= bus("School Volvo", 180, 12)
print(f"Name:{a.name} speed:{a.speed} mileage:{a.mileage}")


````
